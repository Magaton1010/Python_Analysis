# Common Normalization Techniques


**Min-Max Scaling (Normalization):**

Scales features to a specific range (usually 0 to 1).
Formula: (x - min(x)) / (max(x) - min(x))
Pros: Preserves the original distribution shape, easy to interpret.
Cons: Sensitive to outliers, might compress data unnecessarily if outliers are present.

```python
X_min=np.min(xys,0)
X_max=np.max(xys,0)
print(X_min,X_max)
normed_Min_Max_Scaling = (xys-X_min)/(X_max-X_min)
print(normed_Min_Max_Scaling)
```

**Standardization (Z-score Normalization):**

Transforms features to have a mean of 0 and a standard deviation of 1.
Formula: (x - mean(x)) / std(x)
Pros: Handles outliers better than min-max scaling, often preferred when data doesn't follow a normal distribution.
Cons: Doesn't guarantee a specific range, might change the distribution shape.


```python
X_mean= np.mean(xys,0)
X_std= np.std(xys,0)
print(X_mean,X_std)
Z_score_Normalization= (xys-X_mean)/X_std
print(Z_score_Normalization)
```

**Robust Scaling:**

Similar to standardization, but uses median and interquartile range (IQR) instead of mean and standard deviation.
Formula: (x - median(x)) / IQR(x)
Pros: Highly robust to outliers.
Cons: Might compress data more than necessary if outliers aren't a major concern.

```python
X_median = np.median(xys, axis=0)
X_iqr = np.percentile(xys, 75, axis=0) - np.percentile(xys, 25, axis=0)
Robust_Scaling = (xys - X_median) / X_iqr
print(Robust_Scaling)
```

# **Choosing the Right Technique**



If you need features to be in a specific range (e.g., for certain machine learning algorithms): Min-max scaling is a good choice.
If your data contains outliers and you want to minimize their impact: Standardization or robust scaling are preferred.
If you're unsure about the distribution of your data: Standardization is often a safe bet.
