```python
data =pd.read_csv('test-csv.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
happpy=data['happyScore']
income=data['avg_income']
income.max()
```

```python
plt.xlabel('income')
plt.ylabel('happiness')
plt.scatter(income,happpy)

```

```python
data.sort_values(by='avg_income',inplace=True)
data.head()
```

```python
richest=data[data['avg_income']>15000]
richest.iloc[-1]
```

```python
import numpy as np
minimum=data['avg_income'].min()
print(minimum)
print('*********')
print(np.mean(richest['avg_income']))
print(np.mean(minimum))
print(np.mean(data['avg_income']))
```

```python
plt.scatter(richest['avg_income'],richest['happyScore'])
plt.text(richest.iloc[0]['avg_income'],
         richest.iloc[0]['happyScore'],
         richest.iloc[0]['country'])
plt.text(richest.iloc[-1]['avg_income'],
         richest.iloc[-1]['happyScore'],
         richest.iloc[-1]['country'])
plt.xlabel('income')
plt.ylabel('happiness')
```

```python
for k,row in richest.iterrows():
  print(row['country'])
```

```python
plt.scatter(richest['avg_income'],richest['happyScore'])

for k, row in richest.iterrows():
  plt.text(row['avg_income'],
           row['happyScore'],
           row['country'])
plt.xlabel('income')
plt.ylabel('happiness')
```

```python
happpy=data['happyScore']
income=data['avg_income']
ineq= data['income_inequality']
plt.xlabel('income')
plt.ylabel('happiness')
plt.scatter(income,happpy,c=ineq,cmap='viridis')
plt.colorbar( label='income inequality')
```

```python
plt.scatter(income,happpy,s=ineq*10,alpha=0.25)
plt.xlabel('income')
plt.ylabel('happiness')
```
