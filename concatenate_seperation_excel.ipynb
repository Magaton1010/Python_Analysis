# Excel Data Processor

This Python program provides a graphical user interface (GUI) for processing Excel data. It utilizes the `tkinter` library for the GUI and `pandas` for data manipulation. The program allows users to perform two operations on an Excel file:

1. **Separate Values:**
   - Select an Excel file.
   - Enter the column name containing values separated by commas.
   - The program reads the file, splits the values, and creates a new Excel file with each value in a separate row.

2. **Concatenate Values:**
   - Select an Excel file.
   - Enter the column name containing values.
   - The program reads the file, concatenates all values in the specified column with commas, and creates a new Excel file with the concatenated values in a single row.

Users can also save the result of each operation to a new Excel file. The program provides a simple and user-friendly interface for these common Excel data processing tasks.

### Usage:

1. Clone the repository.
2. Run the Python script (`excel_data_processor.py`).
3. Use the "Browse File" button to select an Excel file.
4. Choose the operation: Separate or Concatenate.
5. Enter the column name.
6. Click "Save" to save the result or "Done" to exit the program.
